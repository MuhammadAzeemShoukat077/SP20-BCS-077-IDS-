```python
gender = hello_data['Gender']

l = len(gender)
m=0
f=0
for x in gender:
    if x == "Male":
        m+=1
    elif x == "Female":
        f+=1

pmale=(m/l*100)
pfemale=(f/l*100)

mtlb.rcParams['font.size'] = 20

genders = ['Male', 'Female']
data = [pmale, pfemale]
plt.pie(data, labels = genders,radius = 2.0, autopct='%1.1f%%')
plt.show()
```

```python
results = []

for gpa, gender in zip(hello_data['CGPA'], hello_data['Gender']):
    if gender == 'Male':
        results.append(gpa)
    
plt.hist(results, edgecolor='black', bins=[2, 2.5, 3.0, 3.5, 4])
plt.title('CGPA of Male Students', fontsize=12)
plt.xlabel('CGPA of Males Students', fontsize=10)
plt.ylabel('No. of Students', fontsize=10)
plt.show()

```

```python
marks_of_males = []
marks_of_females = []

for gender, marks in zip(hello_data['Gender'], hello_data['HSSC-1']):
    if gender == 'Male':
        marks_of_males.append(marks)
    else:
         marks_of_females.append(marks)
            
i = 0
x = len(hello_data['HSSC-1']) - len(marks_of_males)
while (i < x):
    marks_of_males.append('NaN')
    i+=1

i = 0
x = len(hello_data['HSSC-1']) - len(marks_of_females)
while (i < x):
    marks_of_females.append('NaN')
    i+=1
    

grades_in_range = hello_data['HSSC-1']
    
plt.scatter(grades_in_range, marks_of_males, marker='x')
plt.scatter(grades_in_range,marks_of_females, marker='o')
plt.xlabel('Marks-Obtained', fontsize=12)
plt.ylabel('Total-Marks', fontsize=12)
plt.title("HSSC-1 marks of all male vs female students", fontsize=12)
plt.legend(["Male" , "Female"])
plt.grid(True)
plt.show()
```
